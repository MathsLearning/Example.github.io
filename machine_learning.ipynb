```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np
import pickle
from sklearn.metrics import r2_score
```

```python
data = pd.read_csv('IceCreamData.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
x = data['Temperature']
y = data['Revenue']
```

```python
x.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.2)
```

```python
model = LinearRegression()
```

```python
print(np.array([x_train]).ndim)
```

```python
train_dim_1 = np.array([x_train]).reshape(-1,1)
```

```python
print(train_dim_1.ndim)
```

```python
model.fit(train_dim_1,y_train)
```

```python
train_dim_2 = np.array([x_test]).reshape(-1,1)
```

```python
print(train_dim_2.ndim)
```

```python
y_predict = model.predict(train_dim_2)
```

```python
print(y_predict)
```

```python
r2_score(y_test,y_predict)
```
